```python
# Basic Tumour Growth Model

#This notebook implements a simple deterministic tumour growth model
#as a starting point for more advanced cancer modelling.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 50, 500)
```

```python
def tumour_growth(t, r=0.2, K=100):
    """
    Logistic tumour growth model.

    r : growth rate
    K : carrying capacity
    """
    return K / (1 + (K - 1) * np.exp(-r * t))
```

```python
tumour_size = tumour_growth(t)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(t, tumour_size, label="Tumour size")
plt.xlabel("Time (days)")
plt.ylabel("Tumour cell population")
plt.title("Deterministic Tumour Growth")
plt.legend()
plt.grid(True)
plt.show()
```
