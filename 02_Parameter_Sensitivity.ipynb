```python
# Tumour Growth Parameter Sensitivity Analysis

#This notebook explores how changes in tumour growth parameters
#affect the size and behaviour of the tumour over time.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tumour_growth(t, r=0.2, K=100):
    """
    Logistic tumour growth model.

    r : growth rate
    K : carrying capacity
    """
    return K / (1 + (K - 1) * np.exp(-r * t))
```

```python
t = np.linspace(0, 50, 500)  # time in days
```

```python
# Define parameter sets
growth_rates = [0.1, 0.2, 0.3]  # slow, medium, fast
carrying_capacities = [80, 100, 120]  # small, medium, large

plt.figure(figsize=(10,6))

# Loop over all combinations
for r in growth_rates:
    for K in carrying_capacities:
        size = tumour_growth(t, r=r, K=K)
        plt.plot(t, size, label=f"r={r}, K={K}")

plt.xlabel("Time (days)")
plt.ylabel("Tumour size")
plt.title("Parameter Sensitivity of Tumour Growth")
plt.legend()
plt.grid(True)
plt.show()
```
